# Transformando Empréstimo em um TAD

Considere a resolução para múltiplos empréstimos.

```c
#include <stdio.h>

typedef struct {
    float s;
    int   n;
    float j;
    int   corrente;
    float p;
} Emprestimo;

Emprestimo novoEmprestimo(float s, int n, float j) {
    Emprestimo umEmprestimo;
    umEmprestimo.s = s;
    umEmprestimo.n = n;
    umEmprestimo.j = j;
    umEmprestimo.corrente = 1;
    umEmprestimo.p = s;
    return umEmprestimo;
}

float proximaParcela(Emprestimo *umEmprestimo) {
    float retorno = umEmprestimo->p;
    if (umEmprestimo->corrente < umEmprestimo->n)
        umEmprestimo->p = umEmprestimo->p + (umEmprestimo->p * (umEmprestimo->j/100));
    else
        umEmprestimo->p = 0;
    (umEmprestimo->corrente)++;
    return retorno;
}

int main() {
    Emprestimo emprestimo1,
               emprestimo2;
    
    emprestimo1 = novoEmprestimo(200, 5, 1);
    emprestimo2 = novoEmprestimo(500, 7, 2);
    
    float p1 = proximaParcela(&emprestimo1),
          p2 = proximaParcela(&emprestimo2);
    
    int i = 1;
    while (p1 > 0 || p2 > 0) {
        if (p1 > 0)
            printf("Emprestimo 1: parcela %d eh %3.2f\n", i, p1);
        if (p2 > 0)
            printf("Emprestimo 2: parcela %d eh %3.2f\n", i, p2);
        p1 = proximaParcela(&emprestimo1);
        p2 = proximaParcela(&emprestimo2);
        i++;
    }   
}
```

# Exercício

Modifique este código (ou um código que você mesmo tenha feito) para que a estrutura (struct) possa representar um Tipo Abstrato de Dados (TAD), em que ela contenha não apenas os valores de estado,como também operações executadas sobre eles. As operações (funções em C) também devem fazer parte do struct do mesmo modo que os campos de dados.

```c
#include <stdio.h>

typedef struct _ {
    float S;
    int N;
    float J;
    float ultimaParcela;
    int numParcela;
    void (*proximaParcela)(int numEmprestimo);
    void (*printParcela)(int numEmprestimo);
} Emprestimo;

// estado
Emprestimo emprestimo[100];
int posicaoDisponivel = 0;
//#######

void proximaParcela(int numEmprestimo) {
    if (emprestimo[numEmprestimo - 1].numParcela == emprestimo[numEmprestimo - 1].N) return;
    emprestimo[numEmprestimo - 1].ultimaParcela = emprestimo[numEmprestimo - 1].ultimaParcela * (1 + emprestimo[numEmprestimo - 1].J / 100);
    emprestimo[numEmprestimo - 1].numParcela++;
}

void printParcela(int numEmprestimo) {
    printf("%dº parcela:%0.2f  \n",emprestimo[numEmprestimo - 1].numParcela, emprestimo[numEmprestimo - 1].ultimaParcela);
}

void initEmprestimo(float S, int N, float J) {
    
    Emprestimo starter;
    emprestimo[posicaoDisponivel] = starter;
    
    emprestimo[posicaoDisponivel].S = S;
    emprestimo[posicaoDisponivel].N = N;
    emprestimo[posicaoDisponivel].J = J;
    emprestimo[posicaoDisponivel].ultimaParcela = S;
    emprestimo[posicaoDisponivel].numParcela = 1;
    emprestimo[posicaoDisponivel].proximaParcela = proximaParcela;
    emprestimo[posicaoDisponivel].printParcela = printParcela;
    posicaoDisponivel++;
}

void main(){
    float S1 = 200;
    int N1 = 5;
    float J1 = 1;
    initEmprestimo(S1,N1,J1);
    
    float S2 = 500;
    int N2 = 7;
    float J2 = 2;
    initEmprestimo(S2,N2,J2);
    
    int maiorQtdParcelas = N2;

    for (int i = 1; i <= maiorQtdParcelas; i++) {
        emprestimo[0].printParcela(1);
        emprestimo[0].proximaParcela(1);
        
        emprestimo[1].printParcela(2);
        emprestimo[1].proximaParcela(2);
    }
}
```
